```python
# Naive Bayes Intuition Visualization

import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Dummy Data
spam_text = "buy cheap meds free money lottery prize winner cash now prince nigeria investment urgent"
ham_text = "meeting schedule project deadline team lunch report analysis statistics data science python"

# 2. Create Plot Layout
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# 3. Plot A: Class Balance (Prior Probability)
labels = ['Spam (Prior: 0.4)', 'Ham (Prior: 0.6)']
sizes = [40, 60]

axes[0].pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    colors=['#ff9999', '#66b3ff'],
    startangle=90
)

axes[0].set_title('Step 1: The Priors P(Class)')

# 4. Plot B: Spam Word Cloud (Likelihood P(Word | Spam))
wc_spam = WordCloud(
    width=400,
    height=400,
    background_color='white',
    colormap='Reds'
).generate(spam_text)

axes[1].imshow(wc_spam)
axes[1].axis('off')
axes[1].set_title('Likelihood: P(Word | Spam)')

# 5. Plot C: Ham Word Cloud (Likelihood P(Word | Ham))
wc_ham = WordCloud(
    width=400,
    height=400,
    background_color='white',
    colormap='Blues'
).generate(ham_text)

axes[2].imshow(wc_ham)
axes[2].axis('off')
axes[2].set_title('Likelihood: P(Word | Ham)')


plt.suptitle("Naive Bayes Intuition: Different Words for Different Classes", fontsize=16)


plt.show()
```
